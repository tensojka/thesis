```python
def gram(M):
    return M*M.transpose()

# counter to count number of recursive calls
minimal norm set to 10000000
counter = 0
minimal_norm = 10000000
counter_before = []
counter_heuristic = []
vectors = []


def gram_approach_rec(G, depth, selected_norm, selected_indices, free_indices, row_sum_selected):
    """
    Finds minimal norm in given lattice, call recursive function.
    Generate all possible linear combinations for the first row given with alpha_i coefficients in {0,1}
    :param G: Gram matrix
    :param depth: depth, level
    :param selected_norm: norm of the chosen vectors based on gram elements
    :param selected_indices: chosen indices of chosen vectors
    :param free_indices: not chosen indices, to be selected
    :param row_sum_selected: list of row sums
    :return: return minimal norm found and relevant indices
    """    
    
    global counter, minimal_norm
    
    counter += 1
    # we found shorter vector
    if minimal_norm > selected_norm and selected_norm != 0:
        minimal_norm = selected_norm
    
    # end of recursion
    if len(free_indices) == 0:
        return
    
    # iterates over all free indices and calls recursively itself
    for i in range(len(free_indices)):
        
        new_ind = free_indices[i] # choose new index, first from free_indices 
        updated_free_indices = free_indices[i+1:] # update free_indices
        
        # all g_ij items we add to the norm if we choose this vector 
        new_col_row = [G[ind, new_ind] for ind in selected_indices] 
        
        # update selected_norm by adding norm of the vector = g_ii and new_col_row twice
        selected_norm += G[new_ind, new_ind] + 2*sum(new_col_row)
        
        test_OK = True
        
        # function to control if we can add this vector (check negative values)
        for k, gij in enumerate(new_col_row):
            row_sum_selected[k] += 2*gij
            
            # i-th from selected indices
            ind_sel_k = selected_indices[k]    
            
            # sum up negative values in free_indices
            negative_values = [G[ind_sel_k, ind] for ind in updated_free_indices if G[ind_sel_k, ind] < 0]
            
            # necessary condition
            if row_sum_selected[k] + 2*sum(negative_values) > 0: 
                test_OK = False
                break
        
        # update sum for row
        row_sum_selected[depth] = G[new_ind, new_ind] + 2*sum(new_col_row)
        
        negative_values = [G[new_ind, ind] for ind in updated_free_indices if G[new_ind, ind] < 0]
        if row_sum_selected[depth] + 2*sum(negative_values) > 0: 
            test_OK = False
            
        # if condition is ok for all chosen vectors -> recursively call itself
        if test_OK:
            updated_selected_indices = selected_indices + [new_ind]
            gram_approach_rec(G, depth=depth+1, selected_norm=selected_norm, selected_indices=updated_selected_indices, free_indices=updated_free_indices, row_sum_selected=row_sum_selected)
        
        for k, gij in enumerate(new_col_row):
            row_sum_selected[k] -= 2*gij
        selected_norm -= G[new_ind, new_ind] + 2*sum(new_col_row)



relation = []
index = 0
for each in G:
    negatives = [each[ind] for ind in range(len(each)) if each[ind] < 0]
    positives = [each[ind] for ind in range(len(each)) if each[ind] > 0 and ind != index]
    if sum(negatives) != 0:
        relation.append((each[index] + 2 * sum(positives)) / abs(2 * sum(negatives)))
    else:
        relation.append((each[index] + 2 * sum(positives)))
    index += 1

relation = sorted(enumerate(relation), key=lambda i: i[1], reverse=True)
new_order = []
for each in relation:
    new_order.append(each[0])

   
```
